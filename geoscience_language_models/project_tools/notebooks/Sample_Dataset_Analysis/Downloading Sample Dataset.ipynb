## Get the list of files from the provided url...

```python
url = URL("https://geoscan.nrcan.gc.ca/tmp/ElementAI/")
```

```python
! pip install urlpath
! pip install bs4
```

```python
from urlpath import URL
from pathlib import Path
```

```python
response = url.get()
```

```python
from bs4 import BeautifulSoup

text = response.text
soup = BeautifulSoup(text, 'html.parser')
print(soup)
    
a_tags = soup.find_all('a')
```

```python
for i, elem in enumerate(a_tags):
    if elem.text == "Parent Directory":
        print(i)
        first_viable_index = i + 1
        break
        
file_tags = a_tags[first_viable_index:]
file_tags
```

```python
# Types of files
import pandas as pd 

file_paths = [Path(file_tag['href']) for file_tag in file_tags]

df = pd.DataFrame({'files': file_paths})
df
```

```python
df['stem'] = df.files.apply(lambda x: x.stem)
df['ext'] = df.files.apply(lambda x: x.suffix)
df['link'] = df.files.apply(lambda x: "https://geoscan.nrcan.gc.ca/tmp/ElementAI/" + str(x))
df
```

## Download the files...

```python
!pip install urllib
```

```python
import urllib
def download_file(download_url, filename):
    response = urllib.request.urlopen(download_url)
    file = open(filename, 'wb')
    file.write(response.read())
    file.close()
    print("Completed")

for irow, row in df.iterrows():
    print(row.link, row.files)
    filename = "/home/stefania/Downloads/NRCAN zips/" + str(row.files)
    if row.ext == '.pdf':
        download_file(row.link, filename)
```
