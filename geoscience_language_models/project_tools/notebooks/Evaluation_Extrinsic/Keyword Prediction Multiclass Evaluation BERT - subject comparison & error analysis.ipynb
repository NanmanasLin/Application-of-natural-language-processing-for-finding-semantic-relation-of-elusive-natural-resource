## Produce a column for precision/recall/f1 for each keyword class

```python
bert_eval_data.iloc[0].transpose()
```

```python
bert_eval_data = bert_eval_data[bert_eval_data['train/eval'] == 'eval']
```

```python
import tqdm
#tqdm.tqdm.pandas()
bert_eval_data_expanded = []
for irow, row in tqdm.tqdm(bert_eval_data.iterrows(), total=bert_eval_data.shape[0]):
    #print(row)
    new_row = row.copy()
    for i in range(len(new_row.eval_support)):
        new_row[f'eval_cat_support_{i}'] = new_row.eval_support[i]
        new_row[f'eval_cat_precision_{i}'] = new_row['eval_sample-precision'][i]
        new_row[f'eval_cat_recall_{i}'] = new_row['eval_sample-recall'][i]
        new_row[f'eval_cat_fb1_{i}'] = new_row['eval_sample-fb1'][i]
        new_row[f'eval_cat_conf_{i}'] = new_row['eval_confusion_matrix'][i]
        new_row[f'eval_cat_tp_{i}'] = new_row['eval_confusion_matrix'][i][1][1]
        new_row[f'eval_cat_fp_{i}'] = new_row['eval_confusion_matrix'][i][0][1]
        new_row[f'eval_cat_tn_{i}'] = new_row['eval_confusion_matrix'][i][0][0]
        new_row[f'eval_cat_fn_{i}'] = new_row['eval_confusion_matrix'][i][1][0]
    #display(new_row)
    bert_eval_data_expanded.append(new_row)
    #break
    
bert_eval_data_expanded =pd.concat(bert_eval_data_expanded, axis=1)
bert_eval_data_expanded
```

```python
bert_eval_data_expanded= bert_eval_data_expanded.transpose()
```

```python
bert_eval_data_expanded
```

```python
bert_eval_data_expanded[['run_dir_full', 'split', 'nrun', 'train/eval', 
                                                        'dataset', 'subject', 'input', 'n_subjects',
       'text_col', 'meta_pipeline', 'bert_model', 'meta_pipeline_binary', 'meta_pipeline_clean',
        'use_class_weights', 'model_shortform', 'model_pipeline','model_dataset', 'tokenizer', 'training_epochs']]
```

```python
for col in bert_eval_data_expanded.filter(regex='eval_cat_(s|p|r|f|t)').columns:
    bert_eval_data_expanded[col] = bert_eval_data_expanded[col].astype(float)
```

```python
bert_eval_data_avg_across_runs = bert_eval_data_expanded.groupby(['run_dir_full', 'split', 'train/eval', 
                                                        'dataset', 'subject', 'input', 'n_subjects',
       'text_col', 'meta_pipeline', 'bert_model', 'meta_pipeline_binary', 'meta_pipeline_clean',
        'use_class_weights', 'model_shortform', 'model_pipeline','model_dataset', 'tokenizer', 'training_epochs']).mean()

bert_eval_data_avg_across_runs
```
