```python

from gensim.test.utils import datapath, get_tmpfile
from gensim.models import KeyedVectors
from gensim.scripts.glove2word2vec import glove2word2vec

glove_file = datapath(f'/nrcan_p2/data/06_models/glove_pretrained/glove.6B.50d.txt')
tmp_file = get_tmpfile("test_word2vec.txt")

_ = glove2word2vec(glove_file, tmp_file)

model = KeyedVectors.load_word2vec_format(tmp_file)
```

```python
def perc_real_words_glove(col):
    return col.str.replace(r'[^\w\s]|_',' ').str.lower().str.split().apply(lambda x: sum([xx in model.vocab for xx in x])/(len(x) if len(x) > 0 else 1))

def perc_real_words_nltk(col):
    return col.str.replace(r'[^\w\s]|_',' ').str.lower().str.split().apply(lambda x: sum([xx in words.words() for xx in x])/(len(x) if len(x) > 0 else 1))

def all_punct(col):
    return col.str.match(r'^([^\w]|_)+$')

def perc_punct(col):
    return col.str.count(r'[^\w\s]') / col.str.replace(r'\s','').str.len()

def all_numeric(col):
    return col.str.match(r'^([^\w]|[0-9])+$')

def perc_numeric_except_punct(col):
    v = col.str.replace(r'[^\w\s]|_',' ')
    return v.str.count(r'[0-9]') / v.str.replace(r'\s','').str.len()
```
