# Ontario Files
We will collect only PDF. We will ignore all no pdf files.
Please also note that the first 2 PDF pages usually have a disclaimer

```python
import pandas as pd
import numpy as np
from os import listdir,environ,rename 
from sys import argv
from os.path import isfile,join,basename
from shutil import rmtree,move
import glob
from bs4 import BeautifulSoup
import os
from zipfile import ZipFile
import requests
from requests_html import HTMLSession,AsyncHTMLSession
import json
```

```python
root_dir = "/nrcan_p2"
data_dir = join(root_dir,"data")
ontario_root_dir=join(data_dir,"01_raw","20210108","ontario")
lists_dir=join(ontario_root_dir,"lists")
```

```python
MAKE_DIRS=False
GET_LINKS=False
DOWNLOAD = False
if MAKE_DIRS:
    os.makedirs(lists_dir)
```

```python
def get_parsed_content(url):
    html_content = requests.get(url, verify=False).text
    soup = BeautifulSoup(html_content, "html.parser")
    return soup

async def get_session_content(url):
    asession = AsyncHTMLSession()
    r = await asession.get(url)
    await r.html.arender()
    html_content = r.html.raw_html
    soup = BeautifulSoup(html_content, "html.parser")
    return soup

def download_articles(article_links,destination_dir,max_length=100):
    for article_link in article_links:
        article_name = article_link.split("/")[-1:][0]
        article_name = article_name[-1*max_length:]
        dest_file = join(destination_dir,article_name)
        print(article_name)
        if not isfile(dest_file):
            r = requests.get(article_link,stream=True, verify=False)
            chunk_size=10000000
            with open (dest_file,"wb") as f:
                for chunk in r.iter_content(chunk_size):
                    f.write(chunk)
                
def get_all_pdf_links(root_html):
    article_links= []
    with open(join(ontario_root_dir,"articles_list.html"),"r") as f:
        article_html = f.read()
        all_links_content = BeautifulSoup(article_html, "html.parser")
    
    for link in all_links_content.find_all('a', href=True):
        if (href := link['href']).startswith("http://www.geologyontario.mndm.gov.on.ca/mndmaccess/mndm_dir.asp"):
            article_content_link = get_parsed_content(href)
            for article_link in article_content_link.find_all('a', href=True):
                if ".pdf" in (link_href := article_link['href']) and not "readme" in link_href.lower():
                    full_link = f'{root_html}{link_href}'
                    article_links.append(full_link)
    return article_links
    
def get_files_count(directory,print_caption=None):
    nb_files = len([name for name in os.listdir(directory) if os.path.isfile(directory +"/"+name)])
    if print_caption:
        print (f'{print_caption}: {nb_files}')
    return nb_files
```

```python
article_links = get_all_pdf_links("http://www.geologyontario.mndm.gov.on.ca")
```

```python
download_articles(article_links,lists_dir)
```

```python
get_files_count(lists_dir,lists_dir)
```
